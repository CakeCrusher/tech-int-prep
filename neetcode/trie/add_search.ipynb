```python
class Node():
    def __init__(self, val = None):
        self.val = val
        self.children = {}
        self.is_final = False
    def insert(self, node):
        self.children[node.val] = node


class WordDictionary(object):

    def __init__(self):
        self.root = Node()
        

    def addWord(self, word):
        """
        :type word: str
        :rtype: None
        """
        curr = self.root
        for char in word:
            if char not in curr.children:
                curr.insert(Node(char))
            curr = curr.children[char]
        curr.is_final = True
        

    def search(self, word):
        """
        :type word: str
        :rtype: bool
        """
        def searchChar(node, i, word):
            if i == len(word)-1:
                if word[i] == '.':
                    for child in node.children:
                        if node.children[child].is_final:
                            return True
                    return False
                if word[i] in node.children:
                    return node.children[word[i]].is_final
                return False
            # print("Char: ", word[i])
            if word[i] in node.children:
                return searchChar(node.children[word[i]], i + 1, word)
            elif word[i] == '.':
                results = []
                for child in node.children:
                    results.append(searchChar(node.children[child], i + 1, word))
                return any(results)
            else:
                return False


        return searchChar(self.root, 0, word)

                    

        


wordDictionary = WordDictionary()
wordDictionary.addWord("bad")
wordDictionary.addWord("dad")
wordDictionary.addWord("mad")
print(wordDictionary.search("pad"))
print(wordDictionary.search("bad"))
print(wordDictionary.search(".ad"))
print(wordDictionary.search("b.."))
```

```python
# depth first search requires iterators inside the recursive function
class TrieNode():
    def __init__(self):
        self.children = {}
        self.is_final = False

class WordDictionary(object):
    def __init__(self):
        self.root = TrieNode()
        
        

    def addWord(self, word):
        """
        :type word: str
        :rtype: None
        """
        curr = self.root
        for char in word:
            if char not in curr.children:
                curr.children[char] = TrieNode()
            curr = curr.children[char]
        curr.is_final = True       
        

    def search(self, word):
        """
        :type word: str
        :rtype: bool
        """
        def dfs(j, node):
            curr = node
            for i in range(j, len(word)):
                if word[i] == ".":
                    for child in curr.children:
                        if dfs(i+1, curr.children[child]):
                            return True
                    return False
                else:
                    if word[i] not in curr.children:
                        return False
                    curr = curr.children[word[i]]
            return curr.is_final
        return dfs(0, self.root)



                    

        


wordDictionary = WordDictionary()
wordDictionary.addWord("bad")
wordDictionary.addWord("dad")
wordDictionary.addWord("mad")
# print(wordDictionary.search("pad"))
print(wordDictionary.search("bad"))
# print(wordDictionary.search(".ad"))
# print(wordDictionary.search("b.."))
```
