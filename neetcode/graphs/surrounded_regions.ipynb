```python
class Solution:
    def solve(self, board) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        coors_visited = set()
        def oob(x,y):
            if x < 0 or x >= len(board[0]) or y < 0 or y >= len(board):
                return True
            return False
        def determine_surrounded(x,y):
            neighbors = [(x,y-1),(x+1,y),(x,y+1),(x-1,y)]
            surrounded_results = []
            surrounded_results.append(((x,y),True))
            for neighbor in neighbors:
                # prevent infinite loop
                if neighbor in coors_visited:
                    continue
                else:
                    coors_visited.add(neighbor)
                

                if oob(neighbor[0],neighbor[1]):
                    surrounded_results.append((neighbor,False))
                else:
                    if board[neighbor[1]][neighbor[0]] == 'X':
                        surrounded_results.append((neighbor,True))
                    else:
                        surrounded_results = surrounded_results + determine_surrounded(neighbor[0],neighbor[1])
            # if any of the items surrounded_results have Fails in the 1 index then return False make this a lambda function
            
            return surrounded_results

            
        for col in range(len(board)):
            for row in range(len(board[0])):
                # if it is 'O' and not visited then use determine_surrounded
                if (row,col) not in coors_visited and board[col][row] == 'O':
                    surrounded_results = determine_surrounded(row,col)
                    if all([item[1] for item in surrounded_results]):
                        for item in surrounded_results:
                            board[item[0][1]][item[0][0]] = 'X'
                    
        
        return board
    
sol = Solution()
print(sol.solve([["X","X","X","X"],["X","O","O","X"],["X","X","O","X"],["X","O","X","X"]]))
```
